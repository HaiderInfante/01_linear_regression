# Generar Variación Porcentual

```python
# Lista de activos
activos = [
    'EURUSD', 'AUDUSD', 'EURCHF', 'USDCAD', 'EURJPY', 'EURGBP', 'GBPJPY',
    'NZDUSD', 'USDJPY', 'USDCHF', 'GBPUSD', 'GBPCHF'
]

# Diccionario para guardar las series de variación porcentual
variaciones = {}

for activo in activos:
    open_col = f'Open_{activo}'
    close_col = f'Close_{activo}'
    var_col = f'VarPct_{activo}'
    
    variaciones[var_col] = ((df[close_col] - df[open_col]) / df[open_col]) * 100

# Crear DataFrame de variaciones
df_variaciones = pd.DataFrame(variaciones)

# Agregar la columna Datetime
df_variaciones.insert(0, 'Datetime', df['Datetime'])

# Ver resultado
df_variaciones
```

```python
df_variaciones.fillna(0, inplace=True)
df_variaciones
```

# Generar Correlaciones Variación Porcentual 15m

```python
# Suponiendo que tu DataFrame se llama df_variaciones
columnas_variaciones = [col for col in df_variaciones.columns if col.startswith('VarPct_')]

# Crear un nuevo DataFrame con las etiquetas 'subió' o 'bajó'
df_etiquetas = df_variaciones.copy()
for col in columnas_variaciones:
    df_etiquetas[col] = df_etiquetas[col].apply(lambda x: 'subió' if x >= 0 else 'bajó')

# Mostrar los primeros valores para verificar
df_etiquetas.head()
```

```python
# Reemplazar etiquetas por 1 (subió) y 0 (bajó)
df_codificado2 = df_etiquetas.replace({'subió': 1, 'bajó': 0})
df_codificado2
```

```python
# Lista de activos con los que se va a comparar (excluyendo EURCHF)
activos = [
    'EURUSD', 'AUDUSD', 'USDCAD', 'EURJPY', 'EURGBP',
    'GBPJPY', 'NZDUSD', 'USDJPY', 'USDCHF', 'GBPUSD', 'GBPCHF'
]

# Filtrar EURCHF a las 00:00:00
df_eurchf = df_codificado2[df_codificado2['Datetime'].dt.time == pd.to_datetime("00:00:00").time()][['Datetime', 'VarPct_EURCHF']]

# Diccionario para guardar correlaciones
resultados = {}

for activo in activos:
    col_activo = f'VarPct_{activo}'

    # Filtrar el activo a las 00:15:00
    df_temp = df_codificado2[df_codificado2['Datetime'].dt.time == pd.to_datetime("00:15:00").time()][['Datetime', col_activo]].copy()

    # Desplazar -15 minutos para alinearlo con EURCHF a las 00:00
    df_temp['Datetime'] = df_temp['Datetime'] - pd.Timedelta(minutes=15)

    # Merge con EURCHF
    df_merge = pd.merge(df_eurchf, df_temp, on='Datetime', how='inner')

    # Calcular correlación si hay datos suficientes
    if len(df_merge) > 0:
        correlacion = df_merge['VarPct_EURCHF'].corr(df_merge[col_activo])
        resultados[activo] = correlacion

# Mostrar resultados ordenados
resultados_ordenados = dict(sorted(resultados.items(), key=lambda x: abs(x[1]), reverse=True))

for activo, corr in resultados_ordenados.items():
    print(f'{activo}: correlación desfase (EURCHF 00:00 vs {activo} 00:15) = {corr:.2f}')
```

```python
from datetime import datetime, timedelta

# Crear lista de horarios en intervalos de 15 minutos
horas = [datetime.strptime("00:00", "%H:%M") + timedelta(minutes=15*i) for i in range(96)]  # 96 bloques de 15 min en 24h

# Resultados globales
resultados_globales = []

for i in range(len(horas)-1):
    h1 = horas[i].time()
    h2 = horas[i+1].time()

    # EURCHF en h1
    df_eurchf = df_codificado2[df_codificado2['Datetime'].dt.time == h1][['Datetime', 'VarPct_EURCHF']]

    for activo in activos:
        col_activo = f'VarPct_{activo}'

        # Activo en h2
        df_temp = df_codificado2[df_codificado2['Datetime'].dt.time == h2][['Datetime', col_activo]].copy()
        df_temp['Datetime'] = df_temp['Datetime'] - pd.Timedelta(minutes=15)  # Alinear con h1

        df_merge = pd.merge(df_eurchf, df_temp, on='Datetime', how='inner')

        if len(df_merge) > 0:
            corr = df_merge['VarPct_EURCHF'].corr(df_merge[col_activo])
            resultados_globales.append({
                'Hora EURCHF': h1.strftime("%H:%M"),
                'Hora Activo': h2.strftime("%H:%M"),
                'Activo': activo,
                'Correlacion': corr
            })

# Convertir a DataFrame
df_resultados = pd.DataFrame(resultados_globales)

# Top 5 correlaciones positivas y negativas (absolutas más fuertes)
top_positivas = df_resultados.sort_values(by='Correlacion', ascending=False).head(5)
top_negativas = df_resultados.sort_values(by='Correlacion').head(5)

print("Top 5 correlaciones positivas más fuertes:")
print(top_positivas)

print("\nTop 5 correlaciones negativas más fuertes:")
print(top_negativas)
```

```python
from datetime import datetime, timedelta
import pandas as pd

# Lista de activos
activos = [
    'EURUSD', 'AUDUSD', 'USDCAD', 'EURJPY', 'EURGBP',
    'GBPJPY', 'NZDUSD', 'USDJPY', 'USDCHF', 'GBPUSD', 'GBPCHF', 'EURCHF'
]

# Crear lista de horarios en intervalos de 15 minutos
horas = [datetime.strptime("00:00", "%H:%M") + timedelta(minutes=15*i) for i in range(96)]  # 96 bloques de 15 min

# Resultados globales
resultados_globales = []

# Loop por cada desfase horario: 00:00 vs 00:15, 00:15 vs 00:30, ...
for i in range(len(horas)-1):
    h1 = horas[i].time()
    h2 = horas[i+1].time()

    for activo_1 in activos:
        for activo_2 in activos:
            if activo_1 == activo_2:
                continue  # Saltar comparación consigo mismo

            col1 = f'VarPct_{activo_1}'
            col2 = f'VarPct_{activo_2}'

            # Filtrar activo_1 en h1
            df_1 = df_codificado2[df_codificado2['Datetime'].dt.time == h1][['Datetime', col1]].copy()

            # Filtrar activo_2 en h2 y alinear -15 min
            df_2 = df_codificado2[df_codificado2['Datetime'].dt.time == h2][['Datetime', col2]].copy()
            df_2['Datetime'] = df_2['Datetime'] - pd.Timedelta(minutes=15)

            # Merge y calcular correlación
            df_merge = pd.merge(df_1, df_2, on='Datetime', how='inner')
            if len(df_merge) > 0:
                corr = df_merge[col1].corr(df_merge[col2])
                resultados_globales.append({
                    'Hora_1': h1.strftime("%H:%M"),
                    'Hora_2': h2.strftime("%H:%M"),
                    'Activo_1': activo_1,
                    'Activo_2': activo_2,
                    'Correlacion': corr
                })

# Convertir a DataFrame
df_resultados = pd.DataFrame(resultados_globales)

# Top 5 correlaciones más fuertes (positivas)
top_positivas = df_resultados.sort_values(by='Correlacion', ascending=False).head(5)

# Top 5 correlaciones más fuertes (negativas)
top_negativas = df_resultados.sort_values(by='Correlacion', ascending=True).head(5)

# Mostrar resultados
print("🔝 Top 5 correlaciones POSITIVAS más fuertes:")
print(top_positivas.to_string(index=False))

print("\n🔻 Top 5 correlaciones NEGATIVAS más fuertes:")
print(top_negativas.to_string(index=False))
```
